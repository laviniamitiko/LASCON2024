#### Data format

```python
f = open('tcs_Jrec=0.01', 'rb'); tcs = pickle.load(f); f.close()
```

```python
trial_no = 20
trial = 0
# number of stimulus orientations to show
stim_no = 33
# range of stimuli
stim_range = np.arange(-160., 160+1, 2*160/(stim_no-1))
label = []


while trial < trial_no:
    trial = trial + 1
    for stim in enumerate(stim_range):
        label.append(stim[1])
```

```python
data = {}
data['spikes'] = np.concatenate(tcs)
data['choices'] = label
```

```python
y = data["choices"]
X = data["spikes"]
```
